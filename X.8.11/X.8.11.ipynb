Явным методом Эйлера получили аналитически связь между $n+1$ и $n$ моментом времени. 

$y_{n+1}=y_n+hz_{n+1}\\z_{n+1}=-100hy_n+(1-101h)z_n$

Для задачи Коши 

$y''+101y'+100y=0\\y(0)=2\\y'(0)=-2$

точное решение $y=2e^{-t}$






```python
import math
import numpy as np
from numpy import linalg as la

N=100
h=0.02
epsilon=h

y = np.zeros(N)
z = np.zeros(N)
t = np.zeros(N)
Y = np.zeros(N)

y[0]=0.5
z[0]=2
t[0]=0
Y[0]=2*math.exp(-t[0])


for i in range(N-1):
  y[i+1]=y[i]+h*z[i]
  z[i+1]=-100*h*y[i]+(1-101*h)*z[i]
  t[i+1]=t[i]+h
  Y[i+1]=2*math.exp(-t[i+1])

max=abs(Y[0]-y[0])
for i in range(N):
  if abs(Y[i]-y[i])>max:
    max=abs(Y[i]-y[i])
print(max)
```

Построим полученные решения (точное и приближенное).

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.plot(t, y)
plt.plot(t, Y)
plt.show()
```

Если проделать то же самое для неявного метода Эйлера

```python
import math
import numpy as np
from numpy import linalg as la

N=100
h=0.5

y = np.zeros(N)
z = np.zeros(N)
t = np.zeros(N)
Y = np.zeros(N)

y[0]=2
z[0]=-2
t[0]=0
Y[0]=2*math.exp(-t[0])
C=1/(100*h**2+101*h+1)



for i in range(N-1):
  y[i+1]=C*((101*h+1)*y[i]+h*z[i])
  z[i+1]=C*(-100*h*y[i]+z[i])
  t[i+1]=t[i]+h
  Y[i+1]=2*math.exp(-t[i+1])

max=abs(Y[0]-y[0])
for i in range(N):
  if abs(Y[i]-y[i])>max:
    max=abs(Y[i]-y[i])
print(max)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.plot(t, y)
plt.plot(t, Y)
plt.show()
```
