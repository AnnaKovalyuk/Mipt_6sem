Построить алгоритм метода пристрелки для вычисления решения следующей нелинейной задачи:

$y” -x\sqrt{y} = 0,~ 0 < x < 1\\y(0)=0,~\int_0^1 y(x)dx=1$.

```python
import math
import numpy as np
from numpy import linalg as la

Y=[]
N=1000
a = np.zeros(N)
da = 5.0/N
A=1

def F(a):
  N=1000
  h=1.0/N
  F=0

  y = np.zeros(N)
  z = np.zeros(N)
  x = np.zeros(N)

  y[0]=0
  z[0]=a
  x[0]=0

  for i in range(N-1):
    y[i+1]=y[i]+h*z[i]
    z[i+1]=z[i]+h*x[i]*math.sqrt(y[i])
    x[i+1]=x[i]+h
  for i in range(N-1):
    F = F +y[i]*h
  F=F-1
  return F

for i in range(N-1):
  a[i+1] = a[i]-F(a[i])*da/(F(a[i]+da)-F(a[i]))

A = a[-1]
print(A)

h=1.0/N

y = np.zeros(N)
z = np.zeros(N)
x = np.zeros(N)

y[0]=0
z[0]=A
x[0]=0

for i in range(N-1):
  y[i+1]=y[i]+h*z[i]
  z[i+1]=z[i]+h*x[i]*math.sqrt(y[i])
  x[i+1]=x[i]+h
```
