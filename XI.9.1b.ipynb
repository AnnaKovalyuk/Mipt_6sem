Построить численно общие решения для следующих дифференциальных уравнений:

б) $y'' + (10 + x^{2})y = xe^{– x},~ 0 < x < 10$.


Т.к. $p(x)=-(10 + x^{2})$ не $>>1$ то общее решение уравнения это 

$y(x)=C_{1}y_{1}(x)+C_{2}y_{2}(x)+y_{0}(x)$

Где $y_{1}(x);~y_{2}(x);~y_{0}(x)$решения задач Коши:

$y_{0}'' + (10 + x^{2})y_{0} = xe^{– x},~ 0 < x < 10\\y_{0}(0)=y'_{0}(0)=0$

$y_{1}'' + (10 + x^{2})y_{1} =0,~ 0 < x < 10\\y_{1}(0)=1\\y'_{1}(0)=0$

$y_{2}'' + (10 + x^{2})y_{2} =0,~ 0 < x < 10\\y_{2}(0)=0\\y'_{2}(0)=1$

Решения которых Явным методом Эйлера получили аналитически связь между $n+1$ и $n$ моментом времени. 

$y_{n+1}=y_n+hz_{n+1}\\z_{n+1}=z_n+h( x_{n}e^{– x_{n}}-(10 + x_{n}^{2})y_{n})$

```python
import math
import numpy as np
from numpy import linalg as la

N=100000
h=5/N
epsilon=h

y0 = np.zeros(N)
z0 = np.zeros(N)
y1 = np.zeros(N)
z1 = np.zeros(N)
y2 = np.zeros(N)
z2 = np.zeros(N)
x = np.zeros(N)

y0[0]=0
z0[0]=0
y1[0]=1
z1[0]=0
y2[0]=0
z2[0]=1
x[0]=0

for i in range(N-1):
  y0[i+1]=y0[i]+h*z0[i]
  z0[i+1]=z0[i]+h*(x[i]*math.exp(x[i])-(10 + x[i]**2)*y0[i])
  y1[i+1]=y1[i]+h*z1[i]
  z1[i+1]=z1[i]+h*(-(10 + x[i]**2)*y1[i])
  y2[i+1]=y2[i]+h*z2[i]
  z2[i+1]=z2[i]+h*(-(10 + x[i]**2)*y2[i])
  x[i+1]=x[i]+h
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```
